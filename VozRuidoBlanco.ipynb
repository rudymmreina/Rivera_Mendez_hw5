```python
import scikits.audiolab as audio
from scipy.fftpack import fft, fftfreq
from __future__ import division
%pylab inline
```

```python
input_signal, sampling_rate, enc = audio.wavread("deutsch.wav")
time_array = arange(0, len(input_signal)/float(sampling_rate), 1/float(sampling_rate))
print len(time_array), len(input_signal)
```

```python
#plot(time_array[0:4000], input_signal[0:4000])
plot(time_array, input_signal)
xlabel("time(s)", fontsize=20)
ylabel("Amplitude", fontsize=20)
```

```python
from scipy.fftpack import fft, fftfreq
fft_v=fft(input_signal)
#duracion en segundos
d=5
freq=fftfreq(len(input_signal),5/len(input_signal))
xlim(-1000,1000)
plot(freq,abs(fft_v))
```

```python
whitef=2000*fft_v/(abs(fft_v))
whitespectrum=whitef**2
scatter(freq,abs(whitespectrum))
```

```python
antiwhite=ifft(whitef)
plot(antiwhite)
output_file = audio.Sndfile("white.wav", 'w', audio.Format('wav'), 1, sampling_rate)
output_file.write_frames(real(antiwhite))
output_file.close()
```
