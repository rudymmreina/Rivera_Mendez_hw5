```python
%pylab inline
import math
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi
from scipy.fftpack import fft, fftfreq, fft2
```

```python
circulos = loadtxt("circulos.dat")
x = circulos[:,0]
y = circulos[:,1]
scatter(x,y,s=0.1)
print len(x)
```

```python
intervalo_x= round(max(x))-round(min(x))
n_points_x=200
h_x=intervalo_x/n_points_x


intervalo_y= round(max(y))-round(min(y))
n_points_y=200
h_y=intervalo_y/n_points_y

A=np.zeros((n_points_x,n_points_y))


for i in range((n_points_x-1)):
    for j in range((n_points_y-1)):
        index = where( (  (x<(-3+((i+1)*h_x) )) & (x>(-3+(i*h_x))) ) &  ( (y>(-3+(j*h_y))) & (y<(-3+((j+1)*h_y))) )    ) 
        A[i][j] = len(index[0])
        
```

```python
plt.imshow(A,cmap="gray")
```

```python
def fourier_2d(imagen):
    im_fft=np.fft.fft2(imagen)
    im_fft[0,0]=0
    im_sfft=np.fft.fftshift(im_fft)
    im_sfft_mag=np.abs(im_sfft)
    im_spec_powr=im_sfft_mag**2
    return im_sfft_mag,im_spec_powr
```
