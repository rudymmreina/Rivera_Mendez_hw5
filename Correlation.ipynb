```python
%pylab inline
import math
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi
from scipy.fftpack import fft, fftfreq, fft2
```

### Cargando los datos 

```python
circulos = loadtxt("circulos.dat")
x = circulos[:,0]
y = circulos[:,1]
scatter(x,y,s=0.1)
xlim(-3,3)
ylim(-3,3)
```

### Convirtiendo las coordenadas (x,y) en una matriz A de 3000X300.Cada intervalo equivale a una distancia de 0.02. Graficando

```python
intervalo_x= round(max(x))-round(min(x))
n_points_x=300
h_x=intervalo_x/n_points_x


intervalo_y= round(max(y))-round(min(y))
n_points_y=300
h_y=intervalo_y/n_points_y

A=np.zeros((n_points_x,n_points_y))


for i in range((n_points_x-1)):
    for j in range((n_points_y-1)):
        index = where( (  (x<(-3+((i+1)*h_x) )) & (x>(-3+(i*h_x))) ) &  ( (y>(-3+(j*h_y))) & (y<(-3+((j+1)*h_y))) )    ) 
        A[i][j] = len(index[0])
        
plt.imshow(A,cmap="gray")
plt.colorbar()
```

### Haciendo la transformada de fourier de la imagen y obteniendo el espectros de potencias

```python

fft_a=fftpack.fft2(A)
fft_fa=np.fft.fftshift(fft_a)
spect=abs(fft_a)**2
back=fftpack.ifft2(fft_a)
imshow(abs(fft_fa)[140:160,140:160])
colorbar()
```

# Comprobación

#### Para comprobar lo dicho anteriormente se prueba con un señal, donde se crea una imagen con varios círculos de radio conocido.  En la siguiente imagen se muestra una composición de 6 círculos de radio de 100 a 105 pixeles

```python
n_points_x=1000

n_points_y=1000


C=np.zeros((n_points_x,n_points_y))


for i in range((n_points_x)):
    for j in range((n_points_y)):
        d=sqrt((200-i)**2+(300-j)**2)
        g=sqrt((500-i)**2+(400-j)**2)
        e=sqrt((400-i)**2+(800-j)**2)
        r=sqrt((800-i)**2+(500-j)**2)
        t=sqrt((300-i)**2+(700-j)**2)
        y=sqrt((700-i)**2+(600-j)**2)
        radio1=100
        radio2=105
        if  (((d)>radio1)&((d)<radio2)) or (((g)>radio1)&((g)<radio2)) or (((e)>radio1)&((e)<radio2)) or (((r)>radio1)&((r)<radio2)) or (((t)>radio1)&((t)<radio2)) or (((y)>radio1)&((y)<radio2)):
            C[i][j] = 1
        else:
            C[i][j] = 0
plt.imshow(C,cmap = cm.Greys_r)
plt.title("Imagen original")
```

### Se realiza el mismo procedimiento, donde se saca la transformada de Fourier de la imagen, luego se calcula el espectro de potencia y por último se calcula la función de correlación (haciendo la inversa de la transformada de Fourier)

```python
fft_c=np.fft.fft2(C)
fft_fc=np.fft.fftshift(fft_c)
spect2=abs(fft_c)**2
back2=fftpack.ifft2(fft_c)
Corr2=fftpack.ifft2(spect2)
Corr2f=np.fft.fftshift(Corr2)
plt.title("Transformada de Fourier")
plt.imshow(abs(fft_fc), cmap = cm.Greys_r)
plt.show()
plt.title("imagen original")
plt.imshow(real(back2), cmap = cm.Greys_r)
plt.show()
fig = figure(1, figsize=(10.5,10.5))
plt.imshow((np.abs(Corr2f)), cmap = cm.Greys_r)
colorbar()

plt.show()

#plt.imshow(real(back2), cmap = cm.Greys_r)
```
